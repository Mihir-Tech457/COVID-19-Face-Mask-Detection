```python
import cv2
import numpy as np
import joblib

# Load Haar Cascade for face detection
haar_data = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the trained SVM model
svm = joblib.load('mask_detection_svm.pkl')  # Ensure this file exists

# Define labels
labels = {0: "No Mask", 1: "Mask"}  # 0 = No Mask, 1 = Mask

# Start webcam capture
capture = cv2.VideoCapture(0)

last_prediction = None  # Keep track of the last prediction

while True:
    flag, img = capture.read()
    if flag:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
        faces = haar_data.detectMultiScale(gray, 1.3, 5)  # Detect faces
        
        if len(faces) == 0 and last_prediction is not None:
            # If no face is detected, maintain the last label on screen
            label = labels[int(last_prediction)]
            color = (0, 255, 0) if last_prediction == 1 else (0, 0, 255)
            cv2.putText(img, label, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
        
        for x, y, w, h in faces:
            face = gray[y:y+h, x:x+w]  # Extract face and keep it grayscale
            face = cv2.resize(face, (50, 50))  # Resize to match training size
            face = face.flatten().reshape(1, -1)  # Convert to 1D array

            # Predict mask or no mask
            pred = svm.predict(face)[0]
            last_prediction = pred  # Update the last prediction
            label = labels[int(pred)]
            
            # Draw rectangle and label
            color = (0, 255, 0) if pred == 1 else (0, 0, 255)  # Green for Mask, Red for No Mask
            cv2.rectangle(img, (x, y), (x+w, y+h), color, 4)
            cv2.putText(img, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
            
            print(label)  # Print result in terminal

        cv2.imshow('Mask Detection', img)

        if cv2.waitKey(2) == 27:  # Press 'Esc' to exit
            break

capture.release()
cv2.destroyAllWindows()
```
